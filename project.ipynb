**Table-1 Orders**

```python
orders=pd.read_csv("olist_orders_dataset.csv",encoding_errors="ignore")
```

```python
orders
```

```python
orders.info()
```

```python
orders.dtypes
```

```python
date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

orders[date_cols] = orders[date_cols].apply(pd.to_datetime)
```

```python
orders.info()
```

```python
# make new column
orders['delivery_days'] = (orders['order_delivered_customer_date'] -orders['order_purchase_timestamp']).dt.days
```

```python
orders
```

```python
orders.duplicated().sum()
```

```python
orders.isnull().sum()
```

```python
orders['order_approved_at'].isnull().sum()
```

```python
orders.head()
```

```python
orders['late_delivery'] = np.where(
    orders['order_delivered_customer_date'] > orders['order_estimated_delivery_date'],
    'Yes',
    'No'
)
```

```python
orders.head()
```

```python
orders.info()
```

```python
orders['delivery_days']=orders['delivery_days'].astype('Int64')
```
